```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# From capture image to model input shape

This section takes the photo captured by streamlit and process it so it has the correct input shape for the model

### Displaying the photo from streamlit

```python
img_2 = cv2.cvtColor(cv2.imread("test.png", cv2.IMREAD_COLOR), cv2.COLOR_BGR2RGB)
plt.imshow(img_2)
```

### Cropping the image to ROI

```python
cropped_img_2 = img_2[200:500,200:500]
plt.imshow(cropped_img_2)
```

### Resizing the image to fit model input shape (128,128,3)

```python
height = 128
width = 128
std_dim = (height, width)
resized_img_2 = cv2.resize(cropped_img_2, std_dim)
plt.imshow(resized_img_2)
```

```python
assert resized_img_2.shape == (128,128,3), "Image is not correctly shaped"
```

### Flipping horizontally the image (if needed)

```python
flipped_img_2 = cv2.flip(resized_img_2, 1)
plt.imshow(flipped_img_2)
```
